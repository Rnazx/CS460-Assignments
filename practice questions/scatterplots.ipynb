```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>(a) Plotting scatter plot for mean at origin</h2>

```python
avg = np.array([0,0])
cov = np.array([[1,0],[0,1]])

x,y = np.random.multivariate_normal(avg,cov,100).T
plt.scatter(x,y)
plt.axis('equal')

```

<h2>(a) Plotting scatter plot for mean at (-1,1)</h2>
<h4> The origin is seen to shift in the scatter diagram</h4>

```python
avg2 = np.array([-1,-1])
x,y = np.random.multivariate_normal(avg2,cov,100).T
plt.scatter(x,y)
plt.axis('equal')
```

<h2>(a) Plotting scatter plot for covariance = [(2,0),(0,2)]</h2>

<h4> The plot is more spread out</h4>

```python
cov2 = np.array([[2,0],[0,2]])

x,y = np.random.multivariate_normal(avg,cov2,100).T
plt.scatter(x,y)
plt.axis('equal')
```

<h2>(a) Plotting scatter plot for covariance = [(1,0.5),(0.5,1)]</h2>

<h4> The plot is skewed along the y=x line. More data points are visible along this line</h4>

```python
cov3 = np.array([[1,0.5],[0.5,1]])

x,y = np.random.multivariate_normal(avg,cov3,100).T
plt.scatter(x,y)
plt.axis('equal')
```

<h2>(a) Plotting scatter plot for covariance = [(1,-0.5),(-0.5,1)]</h2>

<h4> The plot is skewed along the y=-x line. More data points are visible along this line</h4>

```python
cov4 = np.array([[1,-0.5],[-0.5,1]])

x,y = np.random.multivariate_normal(avg,cov4,100).T
plt.scatter(x,y)
plt.axis('equal')
```
