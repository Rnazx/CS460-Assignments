```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the dataset
path = '/home/_rion_/Documents/Semester-07/Machine Learning/tic tac/tic-tac-toe.data'
features = ['top-left-square','top-middle-square','top-right-square', 'middle-left-square', 
            'middle-middle-square','middle-right-square', 'bottom-left-square', 'bottom-middle-square', 'bottom-right-square','Label']
df = pd.read_csv(path,names=features)
df2 = df
df
```

```python
#converting classes to ints
for i in features[:-1]:
    df2[str(i)] = df[str(i)].map({'x': 0.5, 'o': -0.5, 'b': 0})
df2['Label'] = df['Label'].map({'positive': 1, 'negative': -1})
df2

```

```python

arr = df2.values
arr
indices = np.random.permutation(arr.shape[0])
training_idx, test_idx = indices[:80], indices[80:]
training, test = arr[training_idx,:], arr[test_idx,:]
print(training)
print(test)
```

```python
#vanilla perceptron
def vanperceptron(arr,epoch):
    w = np.zeros(len(arr[0])-1)
    b = 0
    for i in range(epoch):
        for j in range(len(arr)):
            a = np.sign(np.dot(w,arr[j][:-1])+b)
            if arr[j][-1]*a <= 0:
                w += (arr[j][-1])*arr[j][:-1]
                b += arr[j][-1]
    return w,b  
def accuracyvan(arr,epochs):
    w,b = vanperceptron(training,epochs)
    count = 0
    for i in range(len(arr)):
        y_pred = (np.sign(np.dot(w,arr[i][:-1])+b))
        if arr[i][-1]*y_pred>=0:
            count+=1
    return(float(count/len(arr)))
accuracyvan(test,5)
```

```python
#voted perceptron
def votperceptron(arr,epoch):
    w = np.zeros(len(arr[0])-1)
    b = 0
    c = 1
    for i in range(epoch):
        C = []
        for j in range(len(arr)):
            a = np.sign(np.dot(w,arr[j][:-1])+b)
            if arr[j][-1]*a <= 0:
                w += (arr[j][-1])*arr[j][:-1]
                b += arr[j][-1]
                C.append([w,b,c])
                c = 1 
            else:
                c += 1
            
    return w,b,C


def accuracyvot(arr,epochs):
    w,b,C = votperceptron(training,epochs)
    count = 0
    for i in range(len(arr)):
        y_pred = 0
        for j in range(len(C)):
            y_pred += C[j][2]*(np.sign(np.dot(C[j][0],arr[i][:-1])+C[j][1]))
        if arr[i][-1]*np.sign(y_pred)>= 0:
            count+=1
    return(float(count/len(arr)))

def accuracyavg(arr,epochs):
    w,b,C = votperceptron(training,epochs)
    count = 0
    for i in range(len(arr)):
        y_pred = 0
        for j in range(len(C)):
            y_pred += C[j][2]*(np.dot(C[j][0],arr[i][:-1])+C[j][1])
        if arr[i][-1]*np.sign(y_pred)>= 0:
            count+=1
    return(float(count/len(arr)))
accuracyvot(test,5)
```

```python
A = np.arange(1,50)
B = []
C = []
D = []
for i in range(1,50):
    B.append(accuracyvan(test,i))
    C.append(accuracyvot(test,i))
    D.append(accuracyavg(test,i))
fig = plt.figure(1, figsize=(4,4))

# Plot vanilla perceptron


ax1 = fig.add_axes([0,0,1,1])
ax1.plot(A,B,label = 'Vanilla Perceptron')
ax1.plot(A,C,label = 'Voted Perceptron')
ax1.plot(A,D,label = 'Average Perceptron')
ax1.legend()
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Number of epochs')
```
